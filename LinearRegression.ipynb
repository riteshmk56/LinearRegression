```python
x=data[['TV','Radio','Newspaper']]
y=data[['Sales']]
```

```python
y
```

```python
                                                                                           # y = b0 + b1x1 + b2x2 + b2x3
! pip install scikit-learn

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape,"Training data size for x")
print(x_test.shape,"Training data size for x")
print(y_train.shape,"Training data size for y")
print(x_test.shape,"Training data size for x")
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model= lm.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# y = b0 + b1x1 + b2x2 + b2x3
b0=2.92588052
b1=0.04692506
b2=0.18475563
b3=-0.00357987
y= b0+b1*17.2+b2*45.9+b3*69.3

```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
new_df = pd.DataFrame()
new_df = x_test
new_df['Actual_Test']= y_test
new_df['predicted_test']= y_pred
new_df
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np

r2 = r2_score(y_test,y_pred)
print('R-squared:',r2)

rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print('RMSE',rmse)
```
